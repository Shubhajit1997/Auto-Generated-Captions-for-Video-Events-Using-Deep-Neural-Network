```python
import math
import operator
import sys
import json
from functools import reduce 
```

```python
test_path = "/content/drive/MyDrive/PROJECTS/VIDEO_CAPTIONING/TENSORFLOW_IMPLEMENTATION/data/testing_data/testing_label.json"
test_caption = "/content/drive/MyDrive/PROJECTS/VIDEO_CAPTIONING/TENSORFLOW_IMPLEMENTATION/data/testing_data/test_greedy.txt"
```

```python
"Calculating bleu score for a all test videos"

from nltk.translate.bleu_score import sentence_bleu

reference = []
candidate = []
score_per_video = []

test_label = json.load(open(test_path,'r'))
test_greedy = open(test_caption,'r')


for count in range(len(test_label)):
  for item in test_label:
    for caption2 in item['caption']:
      caption2 = caption2.rstrip('.')
      caption2 = caption2.split(' ')
      reference.append(caption2)

  for temp in test_greedy:
    id = temp.strip().split(',')[0]
    temp = temp.strip().split(',')[1]
    temp = temp.split(' ')
    candidate = temp
    
    
    score = sentence_bleu(reference,candidate)
    score_per_video.append(score)
```

```python
print('Number of files accessed for calculating bleu score = ',len(score_per_video))
print('\n')
print(score_per_video)
```

```python
average = sum(score_per_video) / len(score_per_video)
print("Average bleu score is ",average)
```
